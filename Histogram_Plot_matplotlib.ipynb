```python
# Data
no = [39, 33, 51, 58, 22, 41, 49, 20, 41, 23, 46, 47, 49, 42, 49, 22, 30, 36, 51, 27, 41, 50, 44, 59,
      53, 30, 38, 37, 13, 11, 17, 55, 54, 30, 54, 14, 33, 24, 40, 28, 18, 46, 26, 40, 13, 57, 35, 17,
      26, 55]
l = [10, 20, 30, 40, 50, 60]

# Plotting the histogram
plt.hist(no, color="b", edgecolor="r", bins=l, label="Python")

# Adding vertical line
line = plt.axvline(45, color="black", label="Line")

# Adding title and labels
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("Number")

# Adjusting the legend
plt.legend(loc="upper left")

# Display the plot
plt.show()
```
